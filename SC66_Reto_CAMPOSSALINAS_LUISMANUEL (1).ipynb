```python
# ============================================================
# PROYECTO: ANALÍTICA DE REDES SOCIALES
# RED DE CORREOS ELECTRÓNICOS DE ENRON
# ============================================================

# ============================================================
# 1. IMPORTAR LIBRERÍAS
# ============================================================

import pandas as pd
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt

from google.colab import files
from IPython.display import display

print("Librerías cargadas correctamente.")


# ============================================================
# 2. SUBIR MANUALMENTE EL ARCHIVO .MTX
# ============================================================

print("Sube manualmente el archivo email-enron-only.mtx")
uploaded = files.upload()

# Tomamos automáticamente el nombre del archivo subido
nombre_archivo = list(uploaded.keys())[0]

print("Archivo cargado:", nombre_archivo)


# ============================================================
# 3. CARGAR EL ARCHIVO EN UN DATAFRAME
# ============================================================

# El archivo .mtx tiene comentarios que empiezan con %
# y contiene dos columnas: nodo que envía y nodo que recibe

df = pd.read_csv(
    nombre_archivo,
    sep=r"\s+",
    comment="%",
    header=None,
    names=["origen", "destino"]
)

print("Primeras filas del DataFrame:")
display(df.head())

print("\nInformación general del DataFrame:")
display(df.info())

print("\nDimensión del DataFrame:")
print(df.shape)


# ============================================================
# 4. REVISIÓN BÁSICA DE LOS DATOS
# ============================================================

print("Número de interacciones en el archivo:", len(df))
print("Valores nulos por columna:")
display(df.isnull().sum())

print("Número de nodos únicos considerando origen y destino:")
nodos_unicos = pd.unique(df[["origen", "destino"]].values.ravel())
print(len(nodos_unicos))


# ============================================================
# 5. CREAR LA RED NO DIRIGIDA
# ============================================================

# Aunque el archivo tiene origen y destino, las instrucciones piden red no dirigida.
# Por eso se usa nx.Graph(), donde la relación se interpreta como una interacción
# entre dos empleados, sin importar quién envió o recibió el correo.

G = nx.from_pandas_edgelist(
    df,
    source="origen",
    target="destino",
    create_using=nx.Graph()
)

print("Red no dirigida creada correctamente.")
print("Número de nodos:", G.number_of_nodes())
print("Número de arcos:", G.number_of_edges())


# ============================================================
# 6. INFORMACIÓN GENERAL DE LA RED
# ============================================================

es_conexa = nx.is_connected(G)
num_componentes = nx.number_connected_components(G)
densidad = nx.density(G)

print("¿La red es conexa?:", es_conexa)
print("Número de componentes conexas:", num_componentes)
print("Densidad de la red:", round(densidad, 4))

# Si la red no fuera conexa, se toma la componente gigante para algunas métricas.
if es_conexa:
    G_analisis = G
else:
    componente_gigante = max(nx.connected_components(G), key=len)
    G_analisis = G.subgraph(componente_gigante).copy()
    print("La red no es conexa, por eso se usa la componente gigante.")
    print("Nodos en componente gigante:", G_analisis.number_of_nodes())
    print("Arcos en componente gigante:", G_analisis.number_of_edges())


# ============================================================
# 7. VISUALIZACIÓN 1: RED BÁSICA
# ============================================================

# Layout fijo con seed para que la gráfica sea reproducible.
pos = nx.spring_layout(G, seed=42, k=0.35)

plt.figure(figsize=(14, 11))

nx.draw_networkx_nodes(
    G,
    pos,
    node_size=90,
    node_color="skyblue",
    alpha=0.85
)

nx.draw_networkx_edges(
    G,
    pos,
    width=0.6,
    alpha=0.25,
    edge_color="gray"
)

nx.draw_networkx_labels(
    G,
    pos,
    font_size=6
)

plt.title("Red no dirigida de correos electrónicos de Enron", fontsize=16)
plt.axis("off")
plt.show()

print("""
Interpretación:
Esta gráfica representa la red general de comunicación por correo electrónico.
Cada nodo representa un empleado y cada arco representa una interacción entre dos empleados.
Como se construyó una red no dirigida, no se distingue quién envió o recibió el correo,
sino únicamente que existió comunicación entre ambos.
""")


# ============================================================
# 8. GRADO DE LOS NODOS
# ============================================================

grados = dict(G.degree())

df_grados = pd.DataFrame(
    list(grados.items()),
    columns=["Nodo", "Grado"]
).sort_values(by="Grado", ascending=False)

print("Nodos con mayor grado:")
display(df_grados.head(10))

print("Nodos con menor grado:")
display(df_grados.tail(10))

nodo_mayor_grado = df_grados.iloc[0]["Nodo"]
mayor_grado = df_grados.iloc[0]["Grado"]

nodo_menor_grado = df_grados.iloc[-1]["Nodo"]
menor_grado = df_grados.iloc[-1]["Grado"]

print("Nodo con mayor grado:", nodo_mayor_grado, "con grado", mayor_grado)
print("Nodo con menor grado:", nodo_menor_grado, "con grado", menor_grado)


# ============================================================
# 9. VISUALIZACIÓN 2: COLOR Y TAMAÑO SEGÚN GRADO
# ============================================================

node_sizes_grado = [grados[n] * 35 + 80 for n in G.nodes()]
node_colors_grado = [grados[n] for n in G.nodes()]

plt.figure(figsize=(14, 11))

nodes = nx.draw_networkx_nodes(
    G,
    pos,
    node_size=node_sizes_grado,
    node_color=node_colors_grado,
    cmap=plt.cm.viridis,
    alpha=0.85
)

nx.draw_networkx_edges(
    G,
    pos,
    width=0.6,
    alpha=0.25,
    edge_color="gray"
)

plt.colorbar(nodes, label="Grado del nodo")
plt.title("Red de Enron: tamaño y color de nodos según su grado", fontsize=16)
plt.axis("off")
plt.show()

print("""
Interpretación:
En esta visualización, los nodos más grandes y con color más intenso son los empleados
con mayor número de conexiones directas. Esto quiere decir que tienen comunicación
con más empleados dentro de la red. Estos nodos pueden ser importantes para difundir
información rápidamente dentro de la empresa.
""")


# ============================================================
# 10. DIÁMETRO DE LA RED
# ============================================================

diametro = nx.diameter(G_analisis)

print("Diámetro de la red analizada:", diametro)

print("""
Interpretación:
El diámetro representa la distancia más larga entre dos nodos, considerando el camino
más corto entre ellos. En palabras simples, indica el máximo número de pasos necesarios
para que un empleado pueda llegar a otro dentro de la red. Un diámetro bajo indica que
la red está relativamente bien conectada.
""")


# ============================================================
# 11. PROMEDIO DE CAMINOS MÁS CORTOS
# ============================================================

promedio_caminos = nx.average_shortest_path_length(G_analisis)

print("Promedio de caminos más cortos:", round(promedio_caminos, 4))

print("""
Interpretación:
El promedio de caminos más cortos indica cuántos pasos se necesitan, en promedio,
para conectar a dos empleados de la red. Si este valor es bajo, significa que la
información puede circular con facilidad porque los empleados están separados por
pocos intermediarios.
""")


# ============================================================
# 12. VERIFICAR SI LA RED ES BIPARTITA
# ============================================================

es_bipartita = nx.is_bipartite(G)

print("¿La red es bipartita?:", es_bipartita)

print("""
Interpretación:
Una red bipartita es aquella donde los nodos pueden dividirse en dos grupos y las
conexiones solo ocurren entre nodos de grupos distintos. En este caso, al ser una red
de correos entre empleados, no necesariamente existe una separación natural en dos
grupos. Por eso, si el resultado es False, significa que la red no cumple con la
estructura bipartita.
""")


# ============================================================
# 13. DISTRIBUCIÓN DEL GRADO
# ============================================================

plt.figure(figsize=(10, 6))

plt.hist(
    list(grados.values()),
    bins=20,
    edgecolor="black",
    alpha=0.75
)

plt.title("Distribución del grado de los nodos", fontsize=15)
plt.xlabel("Grado")
plt.ylabel("Frecuencia")
plt.grid(alpha=0.3)
plt.show()

print("""
Interpretación:
La distribución del grado permite ver cómo se reparten las conexiones entre empleados.
Si muchos nodos tienen grado bajo y pocos tienen grado alto, significa que la red tiene
algunos empleados muy conectados y muchos empleados con participación más limitada.
""")


# ============================================================
# 14. ANÁLISIS DE RESILIENCIA: NODOS DE ARTICULACIÓN
# ============================================================

# Los nodos de articulación son nodos que, al eliminarse, pueden desconectar la red.

nodos_articulacion = list(nx.articulation_points(G))

print("Número de nodos de articulación:", len(nodos_articulacion))
print("Nodos de articulación:")
print(nodos_articulacion)

print("""
Interpretación:
Los nodos de articulación son empleados críticos para la conectividad de la red.
Si uno de estos nodos se elimina, la red puede separarse en diferentes componentes.
Esto significa que estos empleados funcionan como puentes entre grupos de comunicación.
""")


# ============================================================
# 15. VISUALIZACIÓN 3: NODOS DE ARTICULACIÓN
# ============================================================

node_colors_art = [
    "red" if n in nodos_articulacion else "lightblue"
    for n in G.nodes()
]

node_sizes_art = [
    420 if n in nodos_articulacion else 90
    for n in G.nodes()
]

plt.figure(figsize=(14, 11))

nx.draw_networkx_nodes(
    G,
    pos,
    node_color=node_colors_art,
    node_size=node_sizes_art,
    alpha=0.85
)

nx.draw_networkx_edges(
    G,
    pos,
    width=0.6,
    alpha=0.25,
    edge_color="gray"
)

nx.draw_networkx_labels(
    G,
    pos,
    font_size=6
)

plt.title("Nodos de articulación en la red de Enron", fontsize=16)
plt.axis("off")
plt.show()

print("""
Interpretación:
En la gráfica, los nodos rojos son los nodos de articulación. Estos empleados tienen
una posición estructural importante porque su eliminación puede afectar la conexión
entre diferentes partes de la red.
""")


# ============================================================
# 16. ANÁLISIS DE RESILIENCIA: ARCOS PUENTE
# ============================================================

# Los arcos puente son conexiones que, al eliminarse, pueden desconectar la red.

arcos_puente = list(nx.bridges(G))

print("Número de arcos puente:", len(arcos_puente))
print("Arcos puente:")
print(arcos_puente)

print("""
Interpretación:
Los arcos puente son relaciones críticas dentro de la red. Si uno de estos arcos se
elimina, puede romper la comunicación entre ciertos grupos. Esto indica que algunas
conexiones entre empleados son importantes para mantener la red unida.
""")


# ============================================================
# 17. VISUALIZACIÓN 4: ARCOS PUENTE
# ============================================================

edge_colors_bridge = [
    "red" if (u, v) in arcos_puente or (v, u) in arcos_puente else "gray"
    for u, v in G.edges()
]

edge_width_bridge = [
    3 if (u, v) in arcos_puente or (v, u) in arcos_puente else 0.6
    for u, v in G.edges()
]

plt.figure(figsize=(14, 11))

nx.draw_networkx_nodes(
    G,
    pos,
    node_color="lightblue",
    node_size=90,
    alpha=0.85
)

nx.draw_networkx_edges(
    G,
    pos,
    edge_color=edge_colors_bridge,
    width=edge_width_bridge,
    alpha=0.55
)

nx.draw_networkx_labels(
    G,
    pos,
    font_size=6
)

plt.title("Arcos puente en la red de Enron", fontsize=16)
plt.axis("off")
plt.show()

print("""
Interpretación:
En esta gráfica, los arcos rojos representan conexiones puente. Estas conexiones
son relevantes porque su eliminación puede provocar que una parte de la red quede
separada del resto.
""")


# ============================================================
# 18. CENTRALIDAD DE GRADO
# ============================================================

centralidad_grado = nx.degree_centrality(G)

df_centralidad_grado = pd.DataFrame(
    centralidad_grado.items(),
    columns=["Nodo", "Centralidad de grado"]
).sort_values(by="Centralidad de grado", ascending=False)

print("Top 10 nodos por centralidad de grado:")
display(df_centralidad_grado.head(10))

print("""
Interpretación:
La centralidad de grado mide qué tan conectado está directamente un nodo.
Un empleado con alta centralidad de grado tiene interacción directa con muchos empleados.
Esto lo vuelve importante para transmitir información de forma rápida.
""")


# ============================================================
# 19. CENTRALIDAD DE CERCANÍA
# ============================================================

centralidad_cercania = nx.closeness_centrality(G)

df_centralidad_cercania = pd.DataFrame(
    centralidad_cercania.items(),
    columns=["Nodo", "Centralidad de cercanía"]
).sort_values(by="Centralidad de cercanía", ascending=False)

print("Top 10 nodos por centralidad de cercanía:")
display(df_centralidad_cercania.head(10))

print("""
Interpretación:
La centralidad de cercanía mide qué tan cerca está un nodo del resto de la red.
Un empleado con alta cercanía puede llegar a otros empleados usando pocos pasos.
Esto indica una posición eficiente dentro de la estructura de comunicación.
""")


# ============================================================
# 20. CENTRALIDAD DE INTERMEDIACIÓN
# ============================================================

centralidad_intermediacion = nx.betweenness_centrality(G)

df_centralidad_intermediacion = pd.DataFrame(
    centralidad_intermediacion.items(),
    columns=["Nodo", "Centralidad de intermediación"]
).sort_values(by="Centralidad de intermediación", ascending=False)

print("Top 10 nodos por centralidad de intermediación:")
display(df_centralidad_intermediacion.head(10))

print("""
Interpretación:
La centralidad de intermediación mide qué tanto un nodo aparece en los caminos más
cortos entre otros nodos. Un empleado con alta intermediación puede funcionar como
puente entre diferentes grupos. Estos nodos son importantes porque ayudan a conectar
partes de la red que de otra forma estarían más alejadas.
""")


# ============================================================
# 21. CENTRALIDAD EIGENVECTOR
# ============================================================

centralidad_eigenvector = nx.eigenvector_centrality(G, max_iter=1000)

df_centralidad_eigenvector = pd.DataFrame(
    centralidad_eigenvector.items(),
    columns=["Nodo", "Centralidad eigenvector"]
).sort_values(by="Centralidad eigenvector", ascending=False)

print("Top 10 nodos por centralidad eigenvector:")
display(df_centralidad_eigenvector.head(10))

print("""
Interpretación:
La centralidad eigenvector no solo toma en cuenta cuántas conexiones tiene un nodo,
sino también qué tan importantes son los nodos con los que está conectado. Por eso,
un empleado puede tener alta centralidad eigenvector si está conectado con otros
empleados relevantes dentro de la red.
""")


# ============================================================
# 22. CENTRALIDAD PAGERANK
# ============================================================

pagerank = nx.pagerank(G)

df_pagerank = pd.DataFrame(
    pagerank.items(),
    columns=["Nodo", "PageRank"]
).sort_values(by="PageRank", ascending=False)

print("Top 10 nodos por PageRank:")
display(df_pagerank.head(10))

print("""
Interpretación:
PageRank mide la importancia de un nodo considerando tanto sus conexiones como la
importancia de los nodos que se conectan con él. Es útil para identificar empleados
que tienen una posición influyente dentro de la red.
""")


# ============================================================
# 23. TABLA GENERAL DE CENTRALIDADES
# ============================================================

df_centralidades = pd.DataFrame({
    "Nodo": list(G.nodes()),
    "Grado": [grados[n] for n in G.nodes()],
    "Centralidad de grado": [centralidad_grado[n] for n in G.nodes()],
    "Centralidad de cercanía": [centralidad_cercania[n] for n in G.nodes()],
    "Centralidad de intermediación": [centralidad_intermediacion[n] for n in G.nodes()],
    "Centralidad eigenvector": [centralidad_eigenvector[n] for n in G.nodes()],
    "PageRank": [pagerank[n] for n in G.nodes()]
})

df_centralidades = df_centralidades.sort_values(
    by="Centralidad de intermediación",
    ascending=False
)

print("Tabla general de centralidades ordenada por intermediación:")
display(df_centralidades.head(20))


# ============================================================
# 24. VISUALIZACIÓN 5: CENTRALIDAD DE INTERMEDIACIÓN
# ============================================================

node_sizes_betweenness = [
    centralidad_intermediacion[n] * 6000 + 80
    for n in G.nodes()
]

node_colors_betweenness = [
    centralidad_intermediacion[n]
    for n in G.nodes()
]

plt.figure(figsize=(14, 11))

nodes = nx.draw_networkx_nodes(
    G,
    pos,
    node_size=node_sizes_betweenness,
    node_color=node_colors_betweenness,
    cmap=plt.cm.plasma,
    alpha=0.85
)

nx.draw_networkx_edges(
    G,
    pos,
    width=0.6,
    alpha=0.25,
    edge_color="gray"
)

plt.colorbar(nodes, label="Centralidad de intermediación")
plt.title("Red de Enron según centralidad de intermediación", fontsize=16)
plt.axis("off")
plt.show()

print("""
Interpretación:
En esta visualización, los nodos más grandes y con color más intenso tienen mayor
centralidad de intermediación. Estos empleados son relevantes porque se ubican en
rutas de comunicación entre otros empleados. Pueden actuar como conectores o puentes
entre grupos dentro de la empresa.
""")


# ============================================================
# 25. VISUALIZACIÓN 6: CENTRALIDAD EIGENVECTOR
# ============================================================

node_sizes_eigen = [
    centralidad_eigenvector[n] * 3000 + 80
    for n in G.nodes()
]

node_colors_eigen = [
    centralidad_eigenvector[n]
    for n in G.nodes()
]

plt.figure(figsize=(14, 11))

nodes = nx.draw_networkx_nodes(
    G,
    pos,
    node_size=node_sizes_eigen,
    node_color=node_colors_eigen,
    cmap=plt.cm.coolwarm,
    alpha=0.85
)

nx.draw_networkx_edges(
    G,
    pos,
    width=0.6,
    alpha=0.25,
    edge_color="gray"
)

plt.colorbar(nodes, label="Centralidad eigenvector")
plt.title("Red de Enron según centralidad eigenvector", fontsize=16)
plt.axis("off")
plt.show()

print("""
Interpretación:
Esta gráfica muestra qué empleados están conectados con nodos relevantes. La centralidad
eigenvector ayuda a identificar nodos influyentes, no solo por la cantidad de conexiones,
sino por la calidad o importancia de esas conexiones.
""")


# ============================================================
# 26. ANÁLISIS DE ROBUSTEZ DE LA RED
# ============================================================

print("RESUMEN DE ROBUSTEZ")
print("-------------------")
print("¿La red es conexa?:", es_conexa)
print("Número de componentes:", num_componentes)
print("Número de nodos de articulación:", len(nodos_articulacion))
print("Número de arcos puente:", len(arcos_puente))
print("Densidad:", round(densidad, 4))
print("Diámetro:", diametro)
print("Promedio de caminos más cortos:", round(promedio_caminos, 4))

print("""
Interpretación de robustez:
La robustez de una red se refiere a qué tanto puede mantener su conectividad aunque
se eliminen ciertos nodos o arcos. En esta red, los nodos de articulación y los arcos
puente son elementos vulnerables, porque al quitarlos pueden separar partes de la red.

Si la red tiene pocos nodos de articulación y pocos arcos puente, entonces es más
robusta, porque existen caminos alternativos para mantener conectados a los empleados.
Si tiene muchos, entonces la red depende mucho de ciertos empleados o conexiones.

También se debe observar el diámetro y el promedio de caminos más cortos. Un diámetro
moderado y un promedio bajo indican que, en general, los empleados pueden conectarse
en pocos pasos. Esto favorece la circulación de información dentro de la red.
""")


# ============================================================
# 27. RESPUESTAS FINALES PARA EL REPORTE
# ============================================================

print("""
============================================================
RESPUESTAS FINALES DEL PROYECTO
============================================================

1. Carga del archivo y creación de la red:
Se cargó el archivo .mtx en un DataFrame de pandas. Después, usando NetworkX, se
construyó una red no dirigida donde cada nodo representa un empleado y cada arco
representa una interacción por correo electrónico entre dos empleados.

2. Visualización de la red:
Se realizaron varias visualizaciones manipulando el tamaño, color y transparencia
de nodos y arcos. Esto permitió observar mejor la estructura de la red e identificar
visualmente nodos más conectados y conexiones críticas.

3. Nodos con mayor y menor grado:
Los nodos con mayor grado son aquellos empleados que tienen más conexiones directas
dentro de la red. Estos nodos pueden considerarse importantes para la comunicación,
ya que interactúan directamente con muchas personas. Los nodos con menor grado son
empleados más periféricos, con pocas conexiones directas.

4. Diámetro:
El diámetro de la red representa el mayor número de pasos necesarios para conectar
dos nodos mediante el camino más corto. Este valor sirve para entender qué tan
extendida está la red.

5. Promedio de caminos más cortos:
El promedio de caminos más cortos indica cuántos pasos se necesitan en promedio para
conectar a dos empleados. Un valor bajo indica que la información puede circular de
forma relativamente rápida.

6. Red bipartita:
Se verificó si la red es bipartita. Una red bipartita requiere que los nodos puedan
separarse en dos grupos donde solo existan conexiones entre grupos distintos. En una
red de correos entre empleados, normalmente no se cumple esta condición, porque los
empleados pueden comunicarse libremente entre sí.

7. Resiliencia:
Se identificaron nodos de articulación y arcos puente. Estos elementos son importantes
porque su eliminación puede desconectar la red. Por eso representan puntos vulnerables
en la estructura de comunicación.

8. Centralidades:
Se calcularon centralidades de grado, cercanía, intermediación, eigenvector y PageRank.
Cada métrica mide una forma distinta de importancia:
- Grado: número de conexiones directas.
- Cercanía: facilidad para llegar al resto de la red.
- Intermediación: capacidad de funcionar como puente entre otros nodos.
- Eigenvector: conexión con nodos importantes.
- PageRank: importancia considerando la estructura general de conexiones.

9. Conclusión:
El análisis de redes sociales permitió entender la estructura de comunicación dentro
de Enron. Se identificaron empleados con alta conectividad, empleados que funcionan
como puentes y conexiones críticas para mantener unida la red. Esto muestra que no
todos los nodos tienen la misma importancia: algunos destacan por tener muchas
conexiones, mientras que otros son relevantes porque conectan grupos distintos.

Desde el punto de vista de robustez, la red puede considerarse más o menos vulnerable
dependiendo de cuántos nodos de articulación y arcos puente existan. Si la red depende
de pocos nodos clave, entonces puede fragmentarse si estos desaparecen. En cambio,
si existen caminos alternativos, la red será más resistente.
============================================================
""")


# ============================================================
# 28. EXPORTAR RESULTADOS A CSV
# ============================================================

df_grados.to_csv("grados_nodos_enron.csv", index=False)
df_centralidades.to_csv("centralidades_enron.csv", index=False)

print("Archivos CSV generados:")
print("- grados_nodos_enron.csv")
print("- centralidades_enron.csv")

# Descargar los resultados
files.download("grados_nodos_enron.csv")
files.download("centralidades_enron.csv")
```
